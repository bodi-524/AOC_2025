```python
from typing import List
import sys

FILE = 'day2_input.txt'

def read_parts_to_list() -> List[str]:
	parts: List[str] = []
	with open(FILE, "r", encoding="utf-8") as f:
		for line in f:
			parts = line.split(",")

	return parts
```

```python
parts = read_parts_to_list()
invalid_sum = 0
for part in parts:
	a,b = part.split("-")
	for i in range(int(a), int(b)+1):
		i = str(i)
		if len(i) % 2 != 0:
			continue
		else:
			mid = len(i) // 2
			first_half = i[:mid]
			second_half = i[mid:]
			if first_half == second_half:
				invalid_sum += int(i)

print(invalid_sum)
```

```python
from typing import List
import sys

# FILE = 'test.txt'
FILE = 'day2_input.txt'

def read_parts_to_list() -> List[str]:
	parts: List[str] = []
	with open(FILE, "r", encoding="utf-8") as f:
		for line in f:
			parts = line.split(",")

	return parts
```

```python
parts = read_parts_to_list()
invalid_sum = 0

for part in parts:
	a,b = part.split("-")
	for test_subject in range(int(a), int(b)+1): #Loop for each number in the range
		
		test_subject = str(test_subject)
		num_digits = len(test_subject)

		for sub_length in range(1, num_digits//2 + 1): #Loop for each possible sub-length
			if num_digits % sub_length != 0:
				continue
			num_repeats = num_digits // sub_length
			substring = test_subject[:sub_length]
			reconstructed = substring * num_repeats

			if reconstructed == test_subject:
				invalid_sum += int(test_subject)
				break
		
print(invalid_sum)
```
